```python
import datetime as dt
import yfinance as yf
import pandas as pd
```

WE SET DATE RANGE FOR DATA ANALYSIS

```python
end = dt.datetime.now()
start = end - dt.timedelta(days=1259)
start, end
```

WE SELECT STOCKS TO ANALIZE

```python
stockList = ['TSLA', 'AMZN', 'MSFT', 'AAPL', 'GOOG', 'NVDA', 'META']
```

```python
df = yf.download(stockList, start, end)
df.head()
```
